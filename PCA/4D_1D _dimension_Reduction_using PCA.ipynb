```python
# PCA - reducing 4-dimentional data into 1-dim
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# Now take the iris dataset from the scikit learn library
from sklearn import datasets
iris=datasets.load_iris()
iris.data
```

```python
iris.feature_names
```

```python
# take independent feature as 'X'
X=pd.DataFrame(iris.data,columns=iris.feature_names)
X
```

```python
# take independent variable as 'y'
y=iris.target
y
```

# 1--> Normalized data

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_scaler=scaler.fit_transform(X)
X_scaler
```

# Create Covariance Matrix

```python
rows=X_scaler.T
covariance_matrix=np.cov(rows)
covariance_matrix
```

# From Covariance Matrix calculate  eig_values and eig_vectors

```python
eig_values,eig_vectors=np.linalg.eig(covariance_matrix)
```

```python
eig_values,eig_vectors
```

```python
print(eig_values)
```

```python
print(eig_vectors)
```

# Extract only the Eigen_values that is Maximum

```python
print(eig_values)
```

```python
for i in range(len(eig_values)):
    print(eig_values[i]/sum(eig_values))
```

```python
# the oth row contains the highest information gain (72%)
```

# Project the data by using Eigen_vectors

```python
eig_vectors.T
```

```python
eig_vectors.T[0]
```

```python
project_X=X_scaler.dot(eig_vectors.T[0])
```

# Now the dimentionality of the data is reduced

```python
df=pd.DataFrame(project_X,columns=['PC1'])
df
```

```python
df['y_axis']=0.0
df['target']=y
df
```

```python
sns.lmplot('PC1','y_axis',data=df,fit_reg=False,hue='target')
# here the setosa:-0 ,versicolor:-1 ,virginica:-2 ,
```

# Train and Test the Model

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=1)
# X--transformed to X_scaler
# then X_scaler trasnformed to project_X
project_X=pca.fit_transform(X_scaler)
```

```python
# draw the plot
df1=pd.DataFrame(project_X,columns=['PC1'])
df1['y_axis']=0.0
df1['target']=y
df1
```

```python
sns.lmplot('PC1','y_axis',data=df1,fit_reg=False,hue='target')
```

```python
# display dimentionality
print(X_scaler.shape)
print(project_X.shape)
```

```python
#since project_X contains only 1 column
# reshape it as 2D-array
data=project_X.reshape(-1,1)
data
```

```python
# div the data into 2 parts
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(data,y,test_size=0.2)
```

```python
# Now the train the model using SVM
from sklearn.svm import SVC
model=SVC()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
#Find own data prediction
input=np.array([[6.6,3.0,4.4,1.4]])
```

```python
# convert the inputs using StandardScaler
scaled=scaler.transform(input)
scaled
```

```python
# Applt pca and trasnform ( dont use fit)
pca_input=pca.transform(scaled)
```

```python
#make prediction
model.predict(pca_input)
```
