```python
from gensim.models import KeyedVectors
```

```python
# Path to the GloVe file
glove_path = "glove.6B.50d.txt"  # Replace with the correct path if the file is not in the same directory
```

```python
# Load GloVe model
glove_model = KeyedVectors.load_word2vec_format(glove_path, binary=False, no_header=True)
```

```python
# Example: Access a word vector
print("Vector for 'king':", glove_model['king'])
```

```python
# Example: Find similar words
print("Most similar to 'king':", glove_model.most_similar('king'))
```
