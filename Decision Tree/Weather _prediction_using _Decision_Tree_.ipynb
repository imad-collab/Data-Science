```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('weather_forecast.csv')
df
```

# Convert Column to numeric
# Use Label-Encoder

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
# Now apply label encoder on all columns
# outlook---> 0-Overcast,1-Rainy,2-Sunny
#Temp--->0-cool,1-hot,2-mild
#humidity--->0 High,1 Normal
#windy---.0 False,1-True
# Play Cricket---->0-no,1-yes
```

```python
df['outlook_n']=le.fit_transform(df['Outlook'])
df['Temperature_n']=le.fit_transform(df['Temperature'])
df['Humidity_n']=le.fit_transform(df['Humidity'])
df['Windy_n']=le.fit_transform(df['Windy'])
df['Play_n']=le.fit_transform(df['Play'])
```

```python
df
```

```python
df
```

```python
X=df.drop(['Play_n'],axis=1)
y=df['Play_n']
```

# Create DecisionTree Classifier Model

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
model.fit(X,y)
```

```python
#Predict wheather to play cricket or not for the following data:
model.predict([[2,1,0,2]])
```
